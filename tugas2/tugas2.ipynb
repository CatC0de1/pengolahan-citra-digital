```python
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca file gambar
image_path = '../kucing.png'
img = Image.open(image_path)

# Menampilkan gambar asli
plt.figure(figsize=(20, 15))
plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Mengubah gambar menjadi grayscale
img_gray = img.convert('L')

# Menampilkan gambar grayscale
plt.figure(figsize=(20, 15))
plt.subplot(1, 4, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Koordinat untuk memotong gambar (left, upper, right, lower)
crop_box = (70, 50, 90, 70)

# Memotong gambar sesuai dengan koordinat yang diberikan
cropped_img = img_gray.crop(crop_box)

# Menampilkan posisi cropping dengan kotak merah
img_with_box = img_gray.convert("RGB")
draw = ImageDraw.Draw(img_with_box)
draw.rectangle(crop_box, outline="red", width=2)
plt.figure(figsize=(20, 15))
plt.subplot(1, 4, 3)
plt.imshow(img_with_box)
plt.title("Crop Position")
plt.axis("off")
plt.show()
```

```python
# Menampilkan hasil cropping
plt.figure(figsize=(20, 15))
plt.subplot(1, 4, 4)
plt.imshow(cropped_img, cmap='gray')
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
# Mengonversi hasil cropping ke dalam bentuk matriks numpy
img_matrix = np.array(cropped_img)
print(img_matrix)
```

```python
# Menampilkan hasil visualisasi
plt.matshow(img_matrix, cmap="gray")
plt.colorbar()
plt.show()
```
