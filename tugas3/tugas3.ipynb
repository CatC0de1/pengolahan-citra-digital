```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# load image
image = cv2.imread("../kucing.png")

# convert BGR to RGB (for correct color display in Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# display image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
# get image dimensions (M x N)
height, width, channels = image.shape

# display image with spasial
plt.imshow(image_rgb)
plt.xlabel(f"Width: {width} pixels")
plt.ylabel(f"Height: {height} pixels")
plt.show()
```

```python
# convert to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# compute average grayscale intensity
average_gray = np.mean(image_gray)

# display grayscale image with intensity information
plt.imshow(image_gray, cmap="gray")
plt.xlabel(f"Avg Gray Level: {average_gray:.2f}")
```
