```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lightness method
def gray_lightness(img):
  max_rgb = np.max(img, axis=2).astype(np.int16)
  min_rgb = np.min(img, axis=2).astype(np.int16)
  gray = ((max_rgb + min_rgb) // 2).astype(np.uint8)
  return gray
```

```python
# Average method
def gray_average(img):
  gray = np.mean(img, axis=2).astype(np.uint8)
  return gray
```

```python
# Luminosity method
def gray_luminosity(img):
  B, G, R = img[:,:,0], img[:,:,1], img[:,:,2]
  gray = (0.21 * R + 0.71 * G + 0.07 * B).astype(np.uint8)
  return gray
```

```python
# Load image
img = cv2.imread('../../../images/kucing.png')
if img is None:
    print("Gagal memuat gambar!")
    exit()
```

```python
# Transformasi ke grayscale dengan berbagai metode
gray1 = gray_lightness(img)
gray2 = gray_average(img)
gray3 = gray_luminosity(img)
```

```python
# Tampilkan hasil dengan matplotlib
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original')
axs[0, 0].axis('off')

axs[0, 1].imshow(gray1, cmap='gray')
axs[0, 1].set_title('Grayscale - Lightness')
axs[0, 1].axis('off')

axs[1, 0].imshow(gray2, cmap='gray')
axs[1, 0].set_title('Grayscale - Average')
axs[1, 0].axis('off')

axs[1, 1].imshow(gray3, cmap='gray')
axs[1, 1].set_title('Grayscale - Luminosity')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
