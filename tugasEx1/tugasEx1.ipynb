```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# load image
image = cv2.imread("../kucing.png")

# convert BGR to RGB (for correct color display in Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# display image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
x, y = 100, 50
r, g, b = image_rgb[y, x]
print(f"RGB di piksel ({x},{y}): R={r}, G={g}, B={b}")
```

```python
image_no_red = image_rgb.copy()
image_no_green = image_rgb.copy()
image_no_blue = image_rgb.copy()

image_no_red[:, :, 0] = 0  # Kanal merah ada di indeks ke-0 dalam format RGB
image_no_green[:, :, 1] = 0  # Kanal hijau ada di indeks ke-1 dalam format RGB
image_no_blue[:, :, 2] = 0  # Kanal biru ada di indeks ke-2 dalam format RGB
```

```python
plt.imshow(image_no_red)
plt.axis("off")
plt.show()
```

```python
plt.imshow(image_no_green)
plt.axis("off")
plt.show()
```

```python
plt.imshow(image_no_blue)
plt.axis("off")
plt.show()
```

```python
image_red_only = np.zeros_like(image_rgb)
image_green_only = np.zeros_like(image_rgb)
image_blue_only = np.zeros_like(image_rgb)

image_red_only[:, :, 0] = image_rgb[:, :, 0]
image_green_only[:, :, 1] = image_rgb[:, :, 1]
image_blue_only[:, :, 2] = image_rgb[:, :, 2]
```

```python
plt.imshow(image_red_only)
plt.axis("off")
plt.show()

image_red_only_matrik = np.array(image_red_only)
print(image_red_only_matrik)
```

```python
plt.imshow(image_green_only)
plt.axis("off")
plt.show()

image_green_only_matrik = np.array(image_green_only)
print(image_green_only_matrik)
```

```python
plt.imshow(image_blue_only)
plt.axis("off")
plt.show()

image_blue_only_matrik = np.array(image_blue_only)
print(image_blue_only_matrik)
```

```python
plt.subplot(2, 4, 2)
plt.imshow(image_red_only)
plt.title("merah saja")
plt.axis("off")

plt.subplot(2, 4, 3)
plt.imshow(image_green_only)
plt.title("hijau saja")
plt.axis("off")

plt.subplot(2, 4, 4)
plt.imshow(image_blue_only)
plt.title("biru saja")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(image_no_red)
plt.title("tanpa merah")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(image_no_green)
plt.title("tanpa hijau")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(image_no_blue)
plt.title("tanpa biru")
plt.axis("off")
```
